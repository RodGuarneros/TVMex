```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print('setup complete')
```

# Extracting data

```python
IHH_TVAbierta = pd.read_csv('TV abierta.csv', encoding='Latin1', index_col= 'ANIO', parse_dates=True)
print(IHH_TVAbierta.head())

IHH_TVdePaga = pd.read_csv('TV restringida.csv', encoding='Latin1', index_col='FECHA', parse_dates=True)
print(IHH_TVdePaga.head())
```

```python
# Lets display behavior of Transmision Chennels and Programation channels

plt.figure(figsize = (16,6))
sns.lineplot(data=IHH_TVAbierta)
plt.savefig('IHH_TVAbierta.png')
```
